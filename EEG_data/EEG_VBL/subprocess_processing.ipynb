```python
sfreq = clean_file.info["sfreq"]
eeg_cols = [c for c in clean_df.columns if c != "time"]
n_channels = len(eeg_cols)
n_subjects = 17
band_dir = {"alpha": (8, 13), "delta": (0.5, 4), "theta": (4, 8), "beta": (13, 30), "gamma": (30, 45)}
power_all_channels = {i: {} for i in range(n_channels)}

for i, seg in enumerate(segments):
    x = seg[eeg_cols].to_numpy(dtype=float).T # shape: (n_channels, n_samples)
    nperseg = min(int(2 * sfreq), x.shape[1])
    freqs, psd = welch(
        x,
        fs=sfreq,
        window="hann",
        nperseg=nperseg,
        noverlap=nperseg // 2,
        detrend="constant",
        scaling="density",
        axis=1
    )
    for j in range(n_channels):
        for band in band_dir.keys():
            idx = np.logical_and(freqs >= band_dir[band][0], freqs <= band_dir[band][1])
            if power_all_channels[j].get(band):
                power_all_channels[j][band].append(np.mean(psd[j][idx]))
            else:
                power_all_channels[j][band] = [(np.mean(psd[j][idx]))]

power_all_channels_norm = {ch: {} for ch in power_all_channels}
power_all_channels_test = {ch: {} for ch in power_all_channels}
power_all_band = {}
bands = list(band_dir.keys())
chs = list(power_all_channels.keys())

chan_means = {}
for band in bands:
    ch_band_means = []
    for ch in chs:
        vals = np.asarray(power_all_channels[ch][band], dtype=float)  # segment values
        ch_band_means.append(np.nanmean(vals))  # mean over segments
    power_all_band[band] = np.nanmean(ch_band_means)  # mean over channels

print(power_all_band)




    
    
```
